# Elementary Cellular Automata

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Here we will implement the [Wolfram rule](https://mathworld.wolfram.com/ElementaryCellularAutomaton.html) for elementary Cellular automata. Since we have $2^3=8$ possible states for the Elementary Cellular Automaton, we have $2^8 = 256$ possible rules for this automaton.

Wolfram chose to encode this rules as a 8-bit binary number that encodes all possible rules. For example, for the rule 250, we would have $11111010$.

```python
def Elementary_Automata(N,T,n,start=None):
    """Function to construct an Elementary Cellular 
    automata given by a rule N with periodic Boundary 
    Conditions.
    
    Parameters
    -------------------------------------------------
    N: Rule (int)
    T: Number of steps (int)
    n: Number of the lattice (int)
    start(default = None): Starting Configuration of 
                           the lattice. (list)
                           
    Outputs
    -------------------------------------------------
    A: Matrix with the cellular automata evolution
       (np.array)
    
    """
    
    #Transform to a bit-string
    N_Bit = bin(N)[2:]
    
    #Check if the number is an 8-Bit string 
    #if not, you add 0 on the right of the string.
    if len(N_Bit)< 8:
        ad = 8 - len(N_Bit)
        aux = ad*'0'
        N_E_Bit = aux + N_Bit
    else:
        N_E_Bit = N_Bit
    
    # States defined by wolfram
    States = ['111','110','101','100','011','010','001','000']
    #Building a Matrix
    A = np.zeros((T,n))
    #Choose a starting state
    if start == None:
        A[0][np.random.randint(n)] = 1
    else:
        A[0][start] = 1
    for (i,j),_ in np.ndenumerate(A):
        if i>0:
            for B,S in zip(N_E_Bit,States):
                if A[i-1][j-1] == int(S[0]) and A[i-1][j] == int(S[1]) and A[i-1][(j+1)%n] == int(S[2]):
                    A[i][j] = int(B)
    return A
```

```python
plt.imshow(Elementary_Automata(N=90,T=200,n=400,start=200), cmap='Greys');
plt.axis('off');
```
